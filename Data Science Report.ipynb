```python
# --- Libraries ---
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10,6)

# --- Load Data ---
from google.colab import drive
drive.mount('/content/drive')

train = pd.read_csv('/content/drive/MyDrive/DS_project/matches_train.csv')
test = pd.read_csv('/content/drive/MyDrive/DS_project/matches_test.csv')

print("Train shape:", train.shape)
print("Test shape:", test.shape)
```

```python
# --- 1. Summary Statistics ---
print("\n--- Numerical Summary ---")
print(train.describe())

print("\n--- Categorical Summary ---")
print(train.describe(include='object'))
```

```python
# --- 3. Target Variable / Class Balance ---
# Using 'result' as target
train['result'].value_counts().plot(kind='bar', color='skyblue')
plt.title("Class Distribution (Match Result)")
plt.ylabel("Count")
plt.show()
```

```python
# --- 4. Feature Distributions (numerical features) ---
numerical_cols = ['home_team_score', 'away_team_score',
                  'home_team_score_margin', 'away_team_score_margin',
                  'home_team_score_penalties', 'away_team_score_penalties']

train[numerical_cols].hist(bins=15, figsize=(12,8))
plt.suptitle("Feature Distributions")
plt.tight_layout()
plt.show()
```

```python
# --- 5. Correlation Heatmap ---
plt.figure(figsize=(10,8))
sns.heatmap(train[numerical_cols + ['home_team_win','away_team_win','draw']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
